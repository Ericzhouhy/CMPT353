```python
import pandas as pd
```

```python
precipitation = pd.read_csv('precipitation.csv')
precipitation
```

```python
precipitation.insert(loc=3,column='month',value = pd.to_datetime(precipitation['date']).dt.month)
precipitation
```

```python
group = precipitation.groupby(['name','month'])
list(group)
```

```python
sum_pre = group.aggregate({'precipitation': 'sum'}).reset_index()
sum_pre
```

```python
pivot_sum = sum_pre.pivot(index = 'name',columns = 'month',values = 'precipitation')
```
