```python
import pandas as pd
import numpy as py
import os
import matplotlib.pyplot as plt
from scipy import stats
```

```python
data1 = pd.read_csv("data-1.csv")
data2 = pd.read_csv("data-2.csv")
data3 = pd.read_csv("data-3.csv")
data4 = pd.read_csv("data-4.csv")
data5 = pd.read_csv("data-5.csv")
data6 = pd.read_csv("data-6.csv")
```

```python
def get_summary (file,fileName):
    summary = pd.DataFrame(
        {
            'mean_X': file['x'].mean(),
            'mean_Y' : file['y'].mean(),      
            'std_X' : file['x'].std(),
            'std_Y' : file['y'].std(),
            'range_X' : str([file['x'].min(), file['x'].max()]),
            'range_Y' : str([file['y'].min(), file['y'].max()]),
            'corr_coeff' : file.corr()['x']['y']
        },
        index = [fileName]
    )
    return summary
```

```python
def print_plot (file,fileName):
    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)

    plt.title(fileName)

    n1, bins1, patches1 = plt.hist(file['x'],bins='auto',edgecolor='black', alpha=0.5, label='x');
    plt.hist(file['y'],bins=bins1,edgecolor='black', alpha=0.5, label='y');

    plt.legend()

    plt.subplot(1,2,2)
    plt.title(fileName)
    fit = stats.linregress(file['x'],file['y'])
    file['fit'] = file['x']*fit.slope + fit.intercept
    plt.plot(file['x'], file['fit'], 'r-')
    plt.scatter(file['x'],file['y'])
    plt.xlabel('x')
    plt.ylabel('y')
```

```python
print_plot (data1,"data1")
print_plot (data2,"data2")
print_plot (data3,"data3")
print_plot (data4,"data4")
print_plot (data5,"data5")
print_plot (data6,"data6")
```
